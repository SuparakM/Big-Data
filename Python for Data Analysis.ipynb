```python
#Import Python Libraries
import numpy as np
import pandas as pd
```

```python
#1. อ่านไฟล์ข้อมูลชื่อ flights.csv (อยู่บน facebook กลุ่มวิชา)
df = pd.read_csv("flights.csv")
```

```python
#2. แสดงผล 10 แถวแรก
df.head(10)
```

```python
#3. ข้อมูลที่อ่านมาได้มีจำนวนแถวและคอลัมน์เท่าไร
df.shape
```

```python
#4. แสดงผล attribute name ของ distance hour minute ตั้งแต่ แถวที่ 50 ไปอีก 20 แถว
df.loc[50:70,['distance', 'hour', 'minute']]
```

```python
#5. แสดงผลเฉพาะ attribute name ชื่อ flight
df['flight']
```

```python
#6.แสดงผลสถิติแยกตามชื่อสนามบินที่เครื่องบินออก (origin)
df.groupby('origin').describe()
```

```python
#7. แสดงผลค่าเฉลี่ยของระยะทางที่บิน (distance)
df['distance'].mean()
```

```python
#8. จากข้อ 7 ให้แสดงผลค่าเฉลี่ยของระยะทางที่บิน (distance) แยกตามกลุ่มเที่ยวบิน (flight)
df.groupby('flight')[['distance']].mean()
```

```python
#9. ผลที่ได้จากข้อ 6 ให้พิมพ์ค่าเฉลี่ยของการออกไม่ตรงเวลาของสายการบิน (hour)
df.groupby('flight')[['hour']].mean()
```

```python
#10. ต้องการแสดงผลเฉพาะสนามบินปลายทางที่ชื่อ LAS
df[ df['dest'] == 'LAS']
```

```python
#11. เพิ่มเติม ต้องการแสดงผลเฉพาะ attribute dep_delay และ arr_delay ของสนามบินปลายทางที่ชื่อ LAS
df[ df['dest'] == 'LAS'][['dep_delay', 'arr_delay']]
```

```python
#หาค่าเฉลี่ยของ arr_delay 
df['arr_delay'].mean()
```
